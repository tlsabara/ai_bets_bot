# Bot para coletar historico de apostas

Planejo utilizar aqui, pyautogui com o tesseract
O Tesseract não funcionou, então aqui estou apenas coletando os prnts dos valores de crash e apostadores.

```python
# Imports

from pyautogui import locateCenterOnScreen, sleep, screenshot, click, press
from datetime import datetime
from pathlib import Path
import os
```

```python
### Variáveis
ROOT_DIR = Path(os.path.abspath(os.curdir)).parent
# Garante o funcionamento do projeto em qualquer os.

src_images = 'img_src'
dst_images = 'img_dst'
full_screen = True

crash_prefix_save = 'crash_save_{}.png'
crash_save_name = ROOT_DIR / dst_images / crash_prefix_save

fail_save_name_file = 'fail_save_{}.png'
fail_save_name = ROOT_DIR / dst_images / fail_save_name_file

players_prefix_save = 'players_save_{}.png'
players_save_name =  ROOT_DIR / dst_images / players_prefix_save

bet_closed_file = 'apostas_fechadas_full.png'
bet_closed = ROOT_DIR / src_images / bet_closed_file

crashed_file = 'crashed_full.png'
crashed = ROOT_DIR / src_images / crashed_file

reconnect_file = 'reconnect.png'
reconnect = ROOT_DIR / src_images / reconnect_file

insuficient_founds_file = 'insuficient_founds.png'
insuficient_founds = ROOT_DIR / src_images / insuficient_founds_file

no_thanks_file = 'no_thanks.png'
no_thanks = ROOT_DIR / src_images / no_thanks_file

# no_thanks
test_file = 'test.png'
test = ROOT_DIR / src_images / test_file

crash_screenshot_positions = [30,1060, 70, 21] # tentando na barra
players_screenshot_positions = [1487,980, 55, 21] # tentando na barra
```

```python
# Funções

def printd(*args, check: bool = False) -> None:
    if check:
        print(*args)
```

```python
# Monitoramento de status
verify_crashed = None
debug = False
print('Iniciando verificação de apostas...\n')
flag = True
timer = 200000
cont = 1
move = True

print('Iniciando')
while flag:
    if cont % 50 == 0:
        click(30,810)
        press('f5')
        print('Em exec amigo.....',f'{cont}/{timer}')
        cont+=1
    printd('Verificando fechamento das apostas...', check=debug)
    verify_closed = locateCenterOnScreen(str(bet_closed), confidence=0.9)
    if verify_closed is not None:
        id_img = f"{datetime.now().strftime('%Y_%m_%d_%H_%M_%S')}"
        printd('Apostas fechadas!', 'Iniciando verificação do crash', check=debug)
        sleep(1.5)
        players_image = screenshot(region=(players_screenshot_positions))
        players_image.save(str(players_save_name).format(id_img))
        while not verify_crashed:
            printd('Verificando crash...', check=debug)
            verify_crashed = locateCenterOnScreen(str(crashed), confidence=0.9)
            sleep(0.5)
        printd('Crashou...', check=debug)
        sleep(0.5)
        crash_image = screenshot(region=(crash_screenshot_positions))
        sleep(1)
        crash_image.save(str(crash_save_name).format(id_img))
        if move:
            move = !move
            click(40,800)
        else:
            move = !move
            click(30,810)
        verify_crashed = None
        verify_closed = None
        printd('Verificando fechamento das apostas...', check=debug)
        if timer <= cont:
            flag = True
        cont+=1
    # Para falhas de reconectando, apertar F5
    id_img = f"{datetime.now().strftime('%Y_%m_%d_%H_%M_%S')}"
    locate_reconnect = locateCenterOnScreen(str(reconnect), confidence=0.9)
    if locate_reconnect:
        click(30,810)
        img_x, img_y = locateCenterOnScreen(str(reconnect), confidence=0.9)
        reconnect_image = screenshot(region=(img_x, img_y, 137, 46))
        sleep(1)
        reconnect_image.save(str(fail_save_name).format(id_img))
        press('f5')
        print('Reconectando...',f'- {datetime.now()}')
    located_insuficient_founds = locateCenterOnScreen(str(insuficient_founds), confidence=0.9)

    # Para problemas de carteira abaixo do mínimo
    if located_insuficient_founds:
        click(located_insuficient_founds)
        sleep(0.5)
        locate_no_thanks = locateCenterOnScreen(str(no_thanks), confidence=0.9)
        img_x, img_y = locateCenterOnScreen(str(no_thanks), confidence=0.9)
        insuficient_founds_image = screenshot(region=(img_x, img_y, 134, 41))
        insuficient_founds_image.save(str(fail_save_name).format(id_img))
        sleep(1)
        click(locate_no_thanks)
        print('Perguntou do money o safado...',f'- {datetime.now()}')
    sleep(0.5)
```
